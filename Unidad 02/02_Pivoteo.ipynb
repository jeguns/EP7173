# Pivoteo

El pivoteo una técnica de transformación de datos en la que se reorganizan los datos de un formato largo (donde los datos están en una estructura vertical con múltiples filas para cada combinación de clave) a un formato ancho (donde los datos están en una estructura horizontal con una columna para cada valor único de una variable), o viceversa. Esta operación permite ver los datos desde una perspectiva diferente y facilitar el análisis y la visualización.

image.png

```python
import pandas as pd
data = {
    'country': ['x', 'x', 'x', 'y', 'y', 'y', 'z', 'z', 'z'],
    'year': [1960, 1970, 2010, 1960, 1970, 2010, 1960, 1970, 2010],
    'metric': [10, 13, 15, 20, 23, 25, 30, 33, 35]
}

df1 = pd.DataFrame(data)
df1
```

```python
df1a = df1.pivot(index='country', columns='year', values='metric')
df1a.reset_index(inplace=True)
df1a
```

```python
df1b = df1a.melt(id_vars='country', var_name='year', value_name='metric')
df1b
```

## Valores perdidos

```python
import pandas as pd
data = {
    'country': ['x', 'x', 'x', 'y', 'y', 'y', 'z', 'z'],
    'year': [1960, 1970, 2010, 1960, 1970, 2010, 1960, 1970],
    'metric': [10, 13, 15, 20, 23, 25, 30, 33]
}

df2 = pd.DataFrame(data)
df2
```

```python
df2a = df2.pivot(index='country', columns='year', values='metric')
df2a.reset_index(inplace=True)
df2a
```

```python
df2b = df2a.melt(id_vars='country', var_name='year', value_name='metric')
df2b.dropna().sort_values(by='country')
```

## Presencia de duplicados

```python
import pandas as pd
data = {
    'country': ['x', 'x', 'x', 'y', 'y', 'y', 'z', 'z', 'z','z'],
    'year': [1960, 1970, 2010, 1960, 1970, 2010, 1960, 1970, 2010,2010],
    'metric': [10, 13, 15, 20, 23, 25, 30, 33, 35,38]
}

df3 = pd.DataFrame(data)
df3
```

```python
df3a = df3.pivot_table(index='country', columns='year', values='metric', aggfunc='sum')
df3a.reset_index(inplace=True)  # Opcional, para mover 'country' de índice a columna
df3a
```
